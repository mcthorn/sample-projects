

Triangle, pentagonal, and hexagonal numbers are generated by the following formulae:

Triangle 	  	Tn=n(n+1)/2 	  	1, 3, 6, 10, 15, ...

Pentagonal 	  	Pn=n(3n−1)/2 	  	1, 5, 12, 22, 35, ...

Hexagonal 	  	Hn=n(2n−1) 	  	1, 6, 15, 28, 45, ...

It can be verified that T285 = P165 = H143 = 40755.

Find the next triangle number that is also pentagonal and hexagonal.


```python
def tri(n_list):
    f = lambda n: (n * (n+1)) / 2 
    return map(f, n_list)
    
def pent(n_list):
    f =  lambda n: (n * (3*n - 1)) / 2
    return map(f, n_list)
    
def hexa(n_list):
    f = lambda n: n * (2 * n-1) 
    return map(f,n_list)
```

```python
x = range(1,119500)
```

```python
tr = tri(x)
pn = pent(x)
hx = hexa(x)
```

```python
# There's gottta be a way to go through all 3 lists at once

ll = [tr, pn]
ls = [set(l) for l in ll]

su = ls[0]  #union
ssd = ls[0] #symmetric_difference
for s in ls[1:]:
    su = su.union(s)
    ssd = ssd.symmetric_difference(s)
    
result = su.difference(ssd)
print list(result)
```

```python
l2 = [list(result), hx]
ls = [set(l) for l in l2]

su = ls[0]  #union
ssd = ls[0] #symmetric_difference
for s in ls[1:]:
    su = su.union(s)
    ssd = ssd.symmetric_difference(s)
    
result = su.difference(ssd)
print list(result)
```
